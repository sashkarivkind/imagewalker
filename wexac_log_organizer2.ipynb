```python
def vanilla_out_parser(file_name, result_string = 'comitee accuracy:',result_string2 = 'individual accuracy:'):    
    count_test_results = 0
    done_config = False
    accu=-1
    accu2=np.nan
    with open(file_name,'r') as f:
        for line in f:
            if not done_config and line[:6]=='config':
                config_dict_str = line[7:].lstrip()
                done_config = True
            if line[:6] == 'python':
                python_command = line
            if count_test_results > 0:
                accu = line.split(':')[-1]
                accu_list.append(float(accu))
                count_test_results -= 1
            if result_string == line[:len(result_string)]:
                accu = float(line.split(':')[-1])
            if result_string2 == line[:len(result_string2)]:
                accu2 = float(line.split(':')[-1])
            

    # print(accu_list)
    if accu>0:
        config_dict = ast.literal_eval(config_dict_str)
#         print(config_dict)
#         print(np.mean(accu_list))
        config_dict['accuracy'] = np.mean(accu)
        config_dict['accuracy2'] = np.mean(accu2)
#         config_dict['accuracy_std'] = np.std(accu_list)
#         config_dict['accuracy'] = np.mean(accu_list)
        return config_dict
    else:
        return None
```

```python
records = []
for ii,file_name in enumerate(file_names):
    conf=vanilla_out_parser(file_name)
    if conf is not None:
        conf['file_name']=file_name.split('/')[-1]
        conf['file_name_full']=file_name
#         conf['teacher_short']=conf['teacher_net'][-25:]
#         print(conf['teacher_short'])
        records.append(conf)
```
