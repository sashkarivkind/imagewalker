```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
def read_reward_from_logs(run_dir,pos=-1):
    num_template="[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?"

    frame_list=[]
    reward_list=[]
    with open(run_dir+'log.log','r') as f:
        for line in f:
            if 'frame' in line:
                frame_list.append(float(re.findall(num_template,line)[0]))
            if 'running reward' in line:
                reward_list.append(float(re.findall(num_template,line)[pos]))
    return reward_list
```

```python
def read_accuracy_from_logs(run_dir,pos=-1):
    num_template="[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?"

    frame_list=[]
    reward_list=[]
    with open(run_dir+'log.log','r') as f:
        for line in f:
            if 'frame' in line:
                frame_list.append(float(re.findall(num_template,line)[0]))
            if 'accuracy' in line:
                reward_list.append(float(re.findall(num_template,line)[pos]))
    return reward_list #run_syclop_resolution214.py
```
